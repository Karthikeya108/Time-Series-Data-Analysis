## Common Computations

```python
close_px = dataFrame['Adj Close']
```

```python
mavg = pd.rolling_mean(close_px, 30)
mavg[-10:]
```

Returns defined as:

{
    $r_t = \frac{p_t - p_{t-1}}{p_{t-1}}$
}

```python
returns = close_px / close_px.shift(1) - 1
returns.head()
```

```python
avg = returns.mean()
avg
```

```python
var = returns.std()
var
```

## Plotting

## More Data

```python
df = pd.io.data.get_data_yahoo(['AAPL', 'GE', 'IBM', 'KO', 'MSFT', 'PEP'], 
                               start=datetime.datetime(2010, 1, 1), 
                               end=datetime.datetime(2015, 1, 1))['Adj Close']
df.head()
```

```python
rets = df.pct_change()
```

```python
plt.scatter(rets.PEP, rets.KO)
plt.xlabel('Returns PEP')
plt.ylabel('Returns KO')
```

```python
corr = rets.corr()
corr
```

```python
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns);
```

### Relationship of expected returns and the risk

```python
plt.scatter(rets.mean(), rets.std())
plt.xlabel('Expected returns')
plt.ylabel('Risk')
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (20, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```

## Data alignment and dealing with missing values

```python
series_list = []
securities = ['AAPL', 'PEP', 'IBM', 'MSFT']
for security in securities:
    s = pd.io.data.get_data_yahoo(security, start=datetime.datetime(2011, 10, 1), end=datetime.datetime(2013, 1, 1))['Adj Close']
    s.name = security # Rename series to match security name
    series_list.append(s)
```

```python
df = pd.concat(series_list, axis=1)
df.head()
```

```python
df.ix[0, 'AAPL'] = np.nan
df.ix[1, ['PEP', 'IBM']] = np.nan
df.ix[[1, 2, 3], 'MSFT'] = np.nan

df.head()
```

```python
(df.AAPL + df.PEP).head()
```

```python
df.ffill().head()
```
